```python
df['rating'] = pd.to_numeric(df['rating_text'].str.extract(r'(\d+\.?\d*)')[0], errors='coerce')

# Extract last number as number of ratings
df['num_ratings'] = pd.to_numeric(df['rating_text'].str.findall(r'\d+').str[-1], errors='coerce')

df
```

```python
# Select all text columns
text_cols = df.select_dtypes(include="object").columns

for col in text_cols:
    has_issues = (
        df[col].str.startswith(" ") |
        df[col].str.endswith(" ") |
        df[col].str.contains(r"\s{2,}")
    )
    if has_issues.any():   # If any row has the issue
        print(f"Column '{col}' has whitespace problems in {has_issues.sum()} rows")



```

```python
# Select all text columns
text_cols = df.select_dtypes(include='object').columns

# Clean text columns: trim leading/trailing spaces and fix extra internal spaces
df[text_cols] = df[text_cols].apply(lambda x: x.astype(str).str.strip().str.split().str.join(' '))

# Check for any remaining extra spaces
columns_with_spaces = df[text_cols].apply(lambda x: x.str.contains('  ')).any()
print(columns_with_spaces[columns_with_spaces].index.tolist())
```

```python
# Select the columns to standardize
cols_to_standardize = ['language', 'author', 'narrator']

# Convert all text to title case
df[cols_to_standardize] = df[cols_to_standardize].apply(lambda x: x.astype(str).str.lower())
df
```

```python
import string
# Select text columns
text_cols = df.select_dtypes(include='object').columns

# Remove all punctuation from text columns
df[text_cols] = df[text_cols].apply(lambda x: x.astype(str).str.translate(str.maketrans('', '', string.punctuation)))
df
```

```python
# Select all text columns
cols_to_correct = df.select_dtypes(include='object').columns

# Apply trimming and title casing to all text columns
for col in cols_to_correct:
    df[col] = df[col].astype(str).str.strip().str.title()
df
```

```python
df.dtypes
```

```python
# Check data type of price column
if df['price'].dtype == 'object':
    # If price is string, remove currency symbols and commas, then convert to float
    df['price'] = df['price'].replace({'\$':'', ',':''}, regex=True).astype(float)
    print("Price converted from string to float.")
else:
    # If already numeric
    print("Price is already numeric.")
```

```python
# Example column: 'price'
# Step 1: Check if column is numeric
if df['price'].dtype == 'object':
    # Remove non-numeric characters except decimal point
    df['price'] = df['price'].replace(r'[^0-9.]', '', regex=True).astype(float)
    print("Non-numeric characters removed and converted to float.")
else:
    print("Column is already numeric.")
```

```python
# Check and convert rating
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # invalid → NaN

# Check and convert rating_count
df['num_ratings'] = pd.to_numeric(df['num_ratings'], errors='coerce')  # invalid → NaN

# Optional: fill missing or invalid values
df['rating'] = df['rating'].fillna(0)          # or use mean/median
df['num_ratings'] = df['num_ratings'].fillna(0).astype(int)  # or leave as NaN

# Check for any remaining non-numeric values
print(df[['rating','num_ratings']].isna().sum())
```

```python
text_cols = ['author', 'narrator']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()  # Trim and title case
for col in text_cols:
    df[col] = df[col].apply(lambda x: ' '.join(x.split(', ')[::-1]) if ',' in x else x)
```
