```python
import pandas as pd

# Example dataset
data = {
    "Department": ["HR", "HR", "IT", "IT", "Finance", "Finance", "IT"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "George"],
    "Salary": [50000, 52000, 60000, 62000, 58000, 60000, 64000],
    "Bonus": [5000, 6000, 7000, 6500, 5500, 6000, 7200]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Average salary per department
df.groupby("Department")["Salary"].sum()
```

```python
# Total salary per department and bonus level
df.groupby(["Department", "Bonus"])["Salary"].sum()
```

```python
# Multiple statistics at once
df.groupby("Department")["Salary"].agg(["mean", "min", "max", "count"])
```

```python
# Keep only departments where average salary > 55k
filtered = df.groupby("Department").filter(lambda x: x["Salary"].mean() > 55000)
```

```python
filtered
```

```python
# Add a column with average salary per department
df["Dept_Avg_Salary"] = df.groupby("Department")["Salary"].transform("mean")
```

```python
df
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df['total salary'] = df['Salary']+df['Bonus']
```

```python
df
```

```python
df['eligible']= df['total salary']> 80000
```

```python
df['eligible']
```

```python
# Employees with Salary > 60000
df[df["Salary"] > 60000]

# Employees in IT Department
df[df["Department"] == "IT"]
```

```python
# Salary > 60000 AND Department is IT
df[(df["Salary"] > 60000) and (df["Department"] == "IT")]

# Salary < 60000 OR Bonus > 6000
df[(df["Salary"] < 60000) | (df["Bonus"] > 6000)]
```

```python
# Employees in HR or Finance
df[df["Department"].isin(["HR", "Finance"])]
```

```python
# Employees whose names start with 'A'
df[df["Employee"].str.startswith("A")]

# Employees whose names contain 'a'
df[df["Employee"].str.contains("a", case=False)]
```

```python
# Sort by Salary ascending
df.sort_values(by="Salary")

# Sort by Salary descending
df.sort_values(by="Salary", ascending=False)

# Sort by Department, then Salary
df.sort_values(by=["Department", "Salary"], ascending=[True, False])
```

```python
# Rename one or more columns
df = df.rename(columns={
    "Salary": "Base_Salary",
    "Bonus": "Annual_Bonus"
})
print(df.head())
```
