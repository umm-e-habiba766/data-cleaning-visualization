```python
import matplotlib.pyplot as plt
import numpy as np
data = [10,10,25,25,33,33,40,55,55,63,70,70,85,90,90,100,100]

plt.hist(data,bins=30,color='pink',edgecolor="black")
plt.title("Symmetric (No Skewness) with Variation")
plt.show()
df=np.log1p(data)
plt.hist(df,bins=20,color='purple',edgecolor="black")
plt.title("Symmetric (No Skewness) with Variation")
plt.show()

```

```python
import matplotlib.pyplot as plt

# Most values are small, only a few large ones → right tail
data1 = [10,10,15,20,20,25,30,30,35,40,45,50,55,60,100,120,150]

plt.hist(data1,bins=30,color='orange', edgecolor="black")
plt.title("Positive Skewness (Right-Skewed)")
plt.show()
df1=np.log1p(data1)
plt.hist(df1,bins=20,color='lightgreen',edgecolor="black")
plt.title("Positive Skewness (Right-Skewed)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Most values are large, only a few small ones → clearer left tail
data2 = [5, 8, 10, 15, 90, 92, 93, 94, 95, 96, 97, 98, 99, 100, 100, 101, 102, 105]

plt.hist(data2,bins=30,color='yellow', edgecolor="black")
plt.title("Negative Skewness (Left-Skewed)")
plt.show()
df2=np.log1p(data2)
plt.hist(df2,bins=20,color='grey',edgecolor="black")
plt.title("Negative Skewness (Left-Skewed)")
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
labels = ["Apple", "Banana", "Mango","Grapes"]
values = [30, 50, 20,60]
# Create pie chart
plt.pie(values, labels=labels,autopct='%1.1f%%', startangle=90,shadow=True,explode = [0.1, 0, 0.2,0.2],colors = ['pink', 'purple', 'skyblue','yellow'])
plt.title("Fruit Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
ages = [15, 18, 21, 15, 18, 22, 20, 19, 21, 20, 22, 18]

# Create histogram
plt.hist(ages,bins=50,color='skyblue',edgecolor='black',density=True,alpha=0.7,cumulative=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data_symmetrical = np.random.normal(loc=0, scale=1, size=10000)
plt.hist(data_symmetrical, bins=50, color='blue', alpha=0.7)
plt.title('Symmetrical Distribution (No Skewness)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
