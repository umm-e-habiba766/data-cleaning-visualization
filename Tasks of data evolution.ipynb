```python
df["Price_with_Tax"]=df["Price"] * 1.15
df
```

```python
df["Avg_Price_Per_Category"]=df.groupby("Category")["Price"].transform("mean")
df
```

```python
df["Price_Diff_From_Category_Avg"] = df["Price"] - df["Avg_Price_Per_Category"]
df
```

```python
# Calculate total stock per category
df["Category_Total_Stock"] = df.groupby("Category")["Stock"].transform("sum")

# Calculate % contribution of each product
df["Stock_Contribution_Percent"] = (df["Stock"] / df["Category_Total_Stock"]) * 100

# Optional: drop the helper column if you don't need it
df = df.drop(columns=["Category_Total_Stock"])

df
```

```python
# Rank products within each category by price (highest price = rank 1)
df["Price_Rank_In_Category"] = df.groupby("Category")["Price"].rank(ascending=False,method="dense")
df
```

```python
df["Description_First10"] = df["Description"].str[:10]
df[["Description", "Description_First10"]]
```

```python
digital_count = df["Name"].str.contains("Digital", case=False, na=False).sum()
digital_count
```

```python
df["Brand"] = df["Brand"].str.upper()
df["Brand"]
```

```python
memory_products = df[df["Description"].str.contains("memory", case=False, na=False)]
memory_products[["Name", "Description"]]
```

```python
df["Name_Length"] = df["Name"].str.len()
df[["Name", "Name_Length"]]
```

```python
duplicate_ids = df[df["Internal ID"].duplicated()]
duplicate_ids
```

```python
top_expensive = df.nlargest(5, "Price")[["Name", "Price"]]
top_expensive[["Name", "Price"]]
```

```python
lowest_stock = df.nsmallest(5, "Stock")[["Name", "Stock"]]
lowest_stock[["Name", "Stock"]]
```

```python
category_counts = df["Category"].value_counts()
most_products_category = category_counts.idxmax()
most_products_category
```

```python
total_stock_per_brand = df.groupby("Brand")["Stock"].sum()
highest_stock_brand = total_stock_per_brand.idxmax()
highest_stock_brand
```

```python
# Calculate average stock per category
df["Avg_Stock_Per_Category"] = df.groupby("Category")["Stock"].transform("mean")

# Filter products above category average
above_avg_stock = df[df["Stock"] > df["Avg_Stock_Per_Category"]]

above_avg_stock[["Name", "Category", "Stock", "Avg_Stock_Per_Category"]]
```

```python
summary_df = df.groupby("Category").agg(
    Total_Products=("Name", "count"),
    Avg_Price=("Price", "mean"),
    Total_Stock=("Stock", "sum"),
    Min_Price=("Price", "min"),
    Max_Price=("Price", "max")
).reset_index()

summary_df
```

```python
availability_counts = df.groupby("Category")["Availability"].value_counts().unstack(fill_value=0)
availability_counts["InStock_to_PreOrder_Ratio"] = availability_counts.get("in_stock", 0) / availability_counts.get("pre_order", 1)

availability_counts[["InStock_to_PreOrder_Ratio"]]
```

```python
brand_summary = df.groupby("Brand").agg(
    oNumber_of_Products=("Name", "count"),
    oTotal_Stock=("Stock", "sum"),
    oAverage_Price=("Price", "mean")
).reset_index()

brand_summary
```

```python
df["Stock_Status"] = df["Stock"].apply(lambda x: "High Stock" if x > 500 else "Low Stock")
df[["Name", "Stock", "Stock_Status"]]
```
